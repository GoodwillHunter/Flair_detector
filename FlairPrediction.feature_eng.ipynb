```python
import pandas as pd

df = pd.read_csv('data.csv')
df.head()
```

```python
df.columns
```

```python
X = df.drop(["flairs"], axis=1)
X
```

```python
arrX = X.values
```

```python
for i in range(len(arrX)):
    if(arrX[i][5]==True):
        arrX[i][5] = 1
    elif(arrX[i][5]==False):
        arrX[i][5] = 0
```

```python
for i in range(10):
    print(arrX[i][5])
```

```python
X_unscaled = pd.DataFrame(arrX)
X_unscaled
```

```python
X_unscaled.to_csv('X_unscaled.csv')
```

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
df_scaled = pd.DataFrame(min_max_scaler.fit_transform(arrX))
df_scaled.head()
```

```python
from pickle import dump

dump(min_max_scaler, open('scaler.pkl', 'wb'))
```

```python
df_scaled.to_csv('X.csv')
```
